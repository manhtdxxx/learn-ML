```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#
x = np.linspace(-10,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
y4 = x**3

#
fig1,(sub1,sub2) = plt.subplots (nrows=2,ncols=1,figsize=(8,4))
sub1.plot(x,y1,label='Sin') ; sub1.legend(loc=7)
sub2.plot(x,y2,label='Cos') ; sub2.legend(loc=8)

#
fig2,(sub3,sub4) = plt.subplots (nrows=1,ncols=2,figsize=(8,4))
sub3.plot(x,y3,label='x**2') ; sub3.legend(loc=9)
sub4.plot(x,y4,label='x**3') ; sub4.legend(loc=10)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#
x = np.linspace(-10,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
y4 = x**3

#
fig1 = plt.figure(figsize=(10,6))

#
sub1 = fig1.add_subplot(3,1,1)
sub1.plot(x,y1,label = 'Sin') 
sub1.legend(loc=1)

sub1.set_xlabel('Thi gian',fontsize = 15)
sub1.set_ylabel('Bin ',fontsize = 15)

sub1.set_xticks([-10,-5,0,5,10])
sub1.set_xticklabels(['A','B','C','D','E']) # chuyn v xticks trc khi labels v s ticklabels = s ticks
sub1.set_yticks([-2,-1,0,1,2])

sub1.grid()

sub1.set_title('Hnh 1')

#
sub2 = fig1.add_subplot(3,1,3)
sub2.plot(x,y2,label = 'Cos') 
sub2.legend(loc=1)

sub2.set_xlabel('Thi gian',fontsize = 15)
sub2.set_ylabel('Bin ',fontsize = 15)

sub2.set_xticks([-10,-5,0,5,10])
sub2.set_xticklabels(['A','B','C','D','E']) # chuyn v xticks trc khi labels v s ticklabels = s ticks
sub2.set_yticks([-2,-1,0,1,2])

sub2.grid()

sub2.set_title('Hnh 2')

# t tn cho figure, not subplot
fig1.suptitle('ULTRA SOUND WAVE',fontsize = 30)

fig1.savefig('ULTRA SOUND WAVE.png')
```
