```python
from imgtools import *
from PIL import Image
```

```python
dir_path = r'.\img'
img_path = dir_path + '\sunflower03.jpg'
```

```python
img = Image.open(img_path)
img
```

# NH MU RGB --> NH XM = CONVERT()

```python
gray_img = img.convert('L')  # other modes = 'P', '1'
gray_img
```
